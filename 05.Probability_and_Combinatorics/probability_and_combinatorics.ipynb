```python
import numpy as np
import matplotlib.pyplot as plt
from collections import Counter
```

## Probability and combinatorics

```python
dice_throw = np.random.randint(low=1, high=7, size=100)
```

```python
plt.plot(range(1, 101), dice_throw)
plt.xlabel("Trial count")
plt.ylabel("Value")
plt.show()
```

```python
frequencies = [(num, (dice_throw == num).sum() / len(dice_throw)) for num in range(1, 7)]
```

```python
counter = Counter(dice_throw)
```

```python
for item in counter.items():
    print(item[0], item[1] / len(dice_throw))
```

### Plotting the results vs Histogram

```python
plt.hist(dice_throw, bins=11)
plt.xlabel("Outcome")
plt.ylabel("Count")
plt.show()
```

```python
plt.bar([f[0] for f in frequencies], [f[1] for f in frequencies])

plt.xlabel("Outcome")
plt.ylabel("Frequency")
plt.show()
```
