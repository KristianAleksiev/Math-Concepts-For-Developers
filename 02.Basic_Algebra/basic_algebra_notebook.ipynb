# Basic algebra


#### Sum of polynomials

```python
p.polyadd([-8, 5, 2], [-2, 0, 0, 0, 3])
```

$-10 + 5x + 2x^2 + 3x^4 $ => $ 3x^4 + 2x^2+ 5x - 10$

### Multiplication of polynomials

```python
multiplication_result = p.polymul([-8, 5, 2], [-2, 0, 0, 0, 3])
```

$16 - 10x^1 - 4x^2 - 24x^4 + 15x^5 + 6x^6 $  => $6x^6 $

### Using sympy for correct order of monomials

```python
x = sympy.symbols("x")
sympy.Poly(reversed(multiplication_result), x)
```

## Function graphs

```python
def plot_function(f, x_min=-10, x_max=10, n_values=500, show=True):
    x = np.linspace(x_min, x_max, n_values)
    y = f(x)
    plt.plot(x, y)
    if show:
        plt.show()
    
```

```python
plot_function(lambda x: np.sin(x))
```

```python
plot_function(np.sqrt, x_min=0, n_values=500)
```

```python
plot_function(np.log, x_min=0.001, n_values=500)
```

```python
plot_function(lambda x: np.sqrt(1 - x ** 2), x_min=-1, x_max=1, show=False)
plot_function(lambda x: -np.sqrt(1 - x ** 2), x_min=-1, x_max=1, show=False)

plt.gca().set_aspect("equal")
plt.show()
```

## Fundamental Theorem of Algebra - Roots of polynomials

### The number of roots of an equation is determined by the power of the equation

```python
def solve_quadratic_equation(a, b, c):
    D = b * b - 4 * a * c
    print(D >= 0)
    return [
        (-b + math.sqrt(D))/ (2 * a),
        (-b - math.sqrt(D))/ (2 * a)
    ]
```

```python
solve_quadratic_equation(1, 0, 0)
```

```python
solve_quadratic_equation(5, -2, 3)
```

```python
def solve_quadratic_equation_complex(a, b, c):
    D = b * b - 4 * a * c
    print(D >= 0)
    return [
        (-b + cmath.sqrt(D))/ (2 * a),
        (-b - cmath.sqrt(D))/ (2 * a)
    ]

# REAL NUMBERS ARE SUBSET OF COMPLEX NUMBERS
```

```python
solve_quadratic_equation_complex(5, -2, 3)
```

```python
solve_quadratic_equation(1, 0, 0)
```
